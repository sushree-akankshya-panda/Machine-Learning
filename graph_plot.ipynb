```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
series= pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv',header=0, index_col=0)
series.head()
```

```python
series.describe()
```

```python
plt.figure(figsize=(10,5))
plt.plot(series['Passengers'])
plt.show()
```

```python
result=sm.tsa.seasonal_decompose(series, model='additive',period=12)
result.plot()
plt.show()
```
